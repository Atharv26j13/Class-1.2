```python
import random

while 0 == 0:
    a = random.randint(1, 9999)
    b = random.choice(["/", "*", "+", "-"])
    c = random.randint(1, 9999)
    d = random.choice(["/", "*", "+", "-"])
    e = random.randint(1, 9999)

    # Prevent division by zero
    if b == "/" and c == 0:
        c = 1
    if d == "/" and e == 0:
        e = 1

    # Generate expression string
    expression = f"{a} {b} {c} {d} {e}"
    correct_answer = eval(expression)  # Evaluate the expression safely

    print(expression)

    try:
        user_answer = float(input("What is the answer to the question given above? "))
        if abs(user_answer - correct_answer) < 0.0001:  # Allowing small floating-point error
            print("Correct!")
            break  # Exit the loop on the correct answer
        else:
            print("Incorrect. Try again!")
    except ValueError:
        print("Invalid input. Please enter a number.")
```
